```python
from main import versions
from wordcloud import WordCloud
from matplotlib import pyplot as plt
data = {}
wordclouds = {}
```

## Carregador dos arquivos
* para cada versão da biblia:  
    * atribua a variavel file o arquivo que se encontra em txt/versão.txt para somente leitura com codificação UTF8  
    * atribua ao dicionario data na posição versão, os dados que são recebidos da leitura da varivel file  
    * feche o arquivo

```python
for version in versions:
    file = open(f"txt/{version}.txt", 'r', encoding='UTF8')
    data[version] = file.read()
    file.close()
    
```

## Criador das wordclouds e plotagem

```python
for version in data:
    wordclouds[version] = WordCloud(background_color='black').generate(data[version])

    fig, ax = plt.subplots(figsize=(10,2))
    ax.imshow(wordclouds[version], interpolation='bilinear')
    ax.set_axis_off()
    plt.imshow(wordclouds[version])
```
