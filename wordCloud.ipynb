```python
from main import versions
from wordcloud import WordCloud
from matplotlib import pyplot as plt
text = {}
wordclouds = {}
```

## Carregador dos arquivos

```python
for version in versions:
    file = open(f"txt/{version}.txt", 'r', encoding='UTF8')
    text[version] = file.read()
    file.close()
```

## Função que gera o wordcloud

```python
def gerarWordcloudPlot(version, text):
    wordclouds[version] = WordCloud(background_color='black', width=1000, height=1000, ).generate(text)

    fig, ax = plt.subplots(figsize=(10, 10))
    ax.imshow(wordclouds[version], interpolation='bilinear')
    ax.set_axis_off()
    ax.set_title(version.upper(), )
    plt.imshow(wordclouds[version])
```

## Gerador de uma WordCloud com todas as versões

```python
all_text = text['nvi']+ text['acf']+ text['aa']
gerarWordcloudPlot('ALL', all_text)
```

## Iterador para gerar todas as wordclouds das versões

```python
for version in versions:
    text_version = text[version]
    gerarWordcloudPlot(version, text_version)
```
